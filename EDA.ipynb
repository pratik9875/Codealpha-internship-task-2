```python
#Step 2: Load Dataset and Explore Structure
import pandas as pd 
data=pd.read_csv("books_dataset.csv")
df=pd.DataFrame(data)
print(df.head(10))
df.info()
df.describe()
print(df.isnull().sum())
```

```python
df['price']=df['price'].str.replace('√Ç¬£','')
df['price'] = pd.to_numeric(df['price'], errors='coerce')
print(df.head())
```
